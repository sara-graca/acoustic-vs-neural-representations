```python
import pandas as pd
import os 

os.chdir("..")
df = pd.read_csv("data/features/features_acoustic.csv")

# missing per phoneme
print(df.groupby("phoneme")[["F1","F2","F3","f0"]].apply(lambda x: x.isna().mean()).round(3).to_string())

# missing per group
print(df.groupby(["l1_status","gender"])[["F1","F2","F3","f0"]].apply(lambda x: x.isna().mean()).round(3))
```
